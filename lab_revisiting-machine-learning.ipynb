```python
# Starting with a basic descriptive analysis

df.describe().apply(lambda x: round(x,2))
```

```python
# Checking for Null values

dfnull = df.isnull().sum()
Null_columns = dfnull[dfnull>0].index
Null_columns
```

```python
# 2) Exclude the following variables by looking at the definitions. Create a new empty list called drop_list. We will append this list and then drop all the columns in this list later:

# OSOURCE - symbol definitions not provided, too many categories

# ZIP CODE - we are including state already

drop_list = ['OSOURCE','ZIP']
```

```python
# 3) Identify columns that over 85% missing values

dfnull = df.isnull().sum()*100/len(df)
Null_columns = dfnull[dfnull>85].index
Null_columns
```

```python
# 4) Remove those columns from the dataframe

drop_list = drop_list+list(Null_columns)
df_clean = df.drop(drop_list, axis=1)
```

```python
# 5) Reduce the number of categories in the column GENDER. The column should only have either 
# "M" for males, "F" for females, and "other" for all the rest

print(df['GENDER'].value_counts())
```

```python
for i in range(len(df['GENDER'])):
    if (df['GENDER'][i]!='F')&(df['GENDER'][i]!='M'):
        df['GENDER'][i]='other'

print(df['GENDER'].value_counts())
```
